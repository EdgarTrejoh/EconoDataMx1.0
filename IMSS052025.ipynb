```python
# prompt: utilizando la librería seaborn crea un gráfico de barras que permita identificar a partir del dataframe df las columnas Hombres, Mujeres y NoBinarios en el eje Y, en el eje X la columna Periodo únicamente el periodo igual a 31/05/2025, crea un diseño de alto impacto, ejecutivo y colores neutros para identificar cada una de las columnas, inserta el valor correspondiente a cada categoría dando formato de numero separados por comas y sin decimales

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Filter the dataframe for the specified period
df_filtered = df[df['Periodo'] == '31/05/2025'].copy()

# Melt the dataframe to have 'Gender' and 'Value' columns for plotting
df_melted = df_filtered.melt(id_vars=['Periodo'], value_vars=['Hombres', 'Mujeres', 'NoBinarios'], var_name='Genero', value_name='Cantidad')

# Set the style for a clean, executive look
sns.set_style("whitegrid")

# Create the bar plot with a clean color palette
plt.figure(figsize=(8, 5))
barplot = sns.barplot(data=df_melted, x='Genero', y='Cantidad', palette='viridis') # Using a perceptually uniform colormap

# Add titles and labels
plt.title('Distribución por Género al 31/05/2025', fontsize=16, fontweight='bold')
plt.xlabel('')  # No label on x-axis as the gender names are self-explanatory
plt.ylabel('Cantidad', fontsize=12)

# Add value labels on top of the bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():,.0f}',  # Format as comma-separated integer
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 5),
                     textcoords='offset points',
                     fontsize=10)

# Format the y-axis labels with commas and no decimals
barplot.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Improve layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# prompt: utilizando la librería seaborn crea un gráfico de barras que permita identificar a partir del dataframe df las columnas Hombres, Mujeres y NoBinarios en el eje Y, en el eje X la columna Periodo únicamente el periodo igual a 31/05/2025, crea un diseño de alto impacto, ejecutivo y colores neutros para identificar cada una de las columnas, inserta el valor correspondiente a cada categoría dando formato de numero separados por comas y sin decimales, elimina el titulo y la escala del eje Y, no agregues título y que pueda ajustar el ancho de las columnas

import matplotlib.pyplot as plt
# Filter the dataframe for the specified period
df_filtered = df[df['Periodo'] == '31/05/2025'].copy()

# Melt the dataframe to have 'Gender' and 'Value' columns for plotting
df_melted = df_filtered.melt(id_vars=['Periodo'], value_vars=['Hombres', 'Mujeres', 'NoBinarios'], var_name='Genero', value_name='Cantidad')

# Set the style for a clean, executive look with a focus on neutrality
sns.set_theme(style="whitegrid")

# Define a custom neutral color palette
neutral_palette = ["#34495e", "#95a5a6", "#bdc3c7"] # Using colors from the 'Flat UI' palette

# Create the bar plot with a clean color palette and adjusted width
plt.figure(figsize=(6, 4)) # Adjusted figure size for a more compact look
barplot = sns.barplot(data=df_melted, x='Genero', y='Cantidad', palette=neutral_palette, width=1.0) # Reduced bar width

# Remove title and y-axis label and scale
plt.title('')
plt.xlabel('')
plt.ylabel('')
barplot.yaxis.set_visible(False) # Hide the y-axis scale

# Add value labels on top of the bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():,.0f}',  # Format as comma-separated integer
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 5),
                     textcoords='offset points',
                     fontsize=12,
                     color='black') # Make text black for better readability

# Improve layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# prompt: utilizando la librería seaborn crea un gráfico de pastel que permita identificar a partir del dataframe df las columnas Permanentes y Eventuales en el eje Y y en el eje X la columna Periodo únicamente el periodo igual a 31/05/2025, calcular el porcentaje en función de la columna Asegurados, crea un diseño de alto impacto, ejecutivo y colores neutros, inserta una leyenda (annotate) en una parte del gráfico en donde se muestra el total de Asegurados

import pandas as pd
import matplotlib.pyplot as plt
# Filter the DataFrame for the specified period
df_filtered = df[df['Periodo'] == '31/05/2025'].copy()

# Calculate the sum of 'Asegurados' for the legend
total_asegurados = df_filtered['Asegurados'].sum()

# Calculate percentages for 'Permanentes' and 'Eventuales' based on 'Asegurados'
df_filtered['Permanentes_Pct'] = (df_filtered['Permanentes'] / df_filtered['Asegurados']) * 100
df_filtered['Eventuales_Pct'] = (df_filtered['Eventuales'] / df_filtered['Asegurados']) * 100

# Prepare data for the pie chart
pie_data = pd.DataFrame({
    'Category': ['Permanentes', 'Eventuales'],
    'Value': [df_filtered['Permanentes'].sum(), df_filtered['Eventuales'].sum()],
    'Percentage': [df_filtered['Permanentes_Pct'].sum(), df_filtered['Eventuales_Pct'].sum()]
})

# Set a neutral color palette for the pie chart
pie_colors = ["#d9b384", "#a2adba"] # Neutral blue-grey palette

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(pie_data['Value'], labels=pie_data['Category'], autopct='%1.1f%%', startangle=90, colors=pie_colors, wedgeprops={'edgecolor': 'white'})

# Add title
#plt.title('Distribución de Asegurados Permanentes y Eventuales (Mayo 2025)', fontsize=16)

# Add a legend/annotation for the total Asegurados
#plt.text(1.2, 0, f'Total Asegurados: {total_asegurados:.0f}', fontsize=12, bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8))

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.show()
```

```python
# prompt: A partir del dataframe df_ingreso y con la librería Seaborn crea un gráfico de columnas tipo histograma con las columnas Rango y Total

import matplotlib.pyplot as plt
# Crea el gráfico de barras tipo histograma para las columnas 'Rango' y 'Total'
plt.figure(figsize=(12, 7))
barplot_ingreso = sns.barplot(data=df_ingreso, x='Rango', y='Total', palette='viridis')

# Añade títulos y etiquetas
plt.title('Distribución de Ingresos por Rango (Mayo 2025)', fontsize=16, fontweight='bold')
plt.xlabel('Rango de Ingresos', fontsize=12)
plt.ylabel('Número de Asegurados', fontsize=12)

# Rota las etiquetas del eje x si son largas para evitar superposición
plt.xticks(rotation=45, ha='right', fontsize=9)

# Añade etiquetas de valor en la parte superior de las barras
for p in barplot_ingreso.patches:
    barplot_ingreso.annotate(f'{p.get_height():,.0f}',  # Formato con comas
                             (p.get_x() + p.get_width() / 2., p.get_height()),
                             ha='center', va='center',
                             xytext=(0, 5),
                             textcoords='offset points',
                             fontsize=7)

# Formato del eje y para que muestre números enteros con comas
barplot_ingreso.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Mejora el diseño y muestra el gráfico
plt.tight_layout()
plt.show()
```
